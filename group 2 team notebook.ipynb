```python
import glob
import shutil
```

```python
book_list1 = glob.glob('*book1*')
book_list2 = glob.glob('*book2*')
book_list3 = glob.glob('*book3*')
book_list4 = glob.glob('*book4*')
book_list5 = glob.glob('*book5*')
```
